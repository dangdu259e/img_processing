```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
def getHistogramGrap(img, title):
    # using numpy
    # hist: histogram
    # bins: 0-256
    hist, bins = np.histogram(img.flatten(), bins=256, range=[0, 256])
    plt.title(title)
    plt.xlabel('bins')
    plt.ylabel('hist')
    plt.plot(hist)
    plt.show()
```

```python
if __name__ == '__main__':
    path = "./image/test.jpg"
    img_original = cv2.imread(path)
    cv2.imshow("Origin", img_original)
    getHistogramGrap(img_original, "Histogram Original Image")
    # equalization histogram
    img_equalized = cv2.equalizeHist(img_original)
    cv2.imshow("Img Equalized Histogram", img_equalized)
    getHistogramGrap(img_equalized, "Histogram Equalization Image")
```
